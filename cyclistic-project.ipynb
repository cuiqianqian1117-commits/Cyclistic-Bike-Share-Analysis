```python
#Combine 2025 full year data into 1 csv file
import pandas as pd
import glob

files = glob.glob('/kaggle/input/datasets/qianqiancuiapril/cyclist/*.csv')
df_list = [pd.read_csv(f) for f in files]
full_year_data = pd.concat(df_list,ignore_index = True)
```

```python
#check the total row number and column number to confirm the combination is finished
full_year_data.shape
```

```python
#export one version without any calculation
full_year_data.to_csv('/kaggle/working/cyclistic_data_raw.csv', index = False)
```

```python
#added 2 columns: ride_length and day_of_week
#transfer format of ride_length from string to float(seconds) for further calculation
full_year_data['started_at'] = pd.to_datetime(full_year_data['started_at'])
full_year_data['ended_at'] = pd.to_datetime(full_year_data['ended_at'])
full_year_data['ride_length'] = full_year_data['ended_at'] - full_year_data['started_at'] 
full_year_data['ride_length'] = full_year_data['ride_length'].dt.total_seconds()
full_year_data['day_of_week'] = full_year_data['started_at'].dt.weekday + 1
full_year_data[['ride_length', 'day_of_week']]
```

```python
#calculate mean and max of ride_length
#Calculate the mode of day_of_week
print(full_year_data['ride_length'].mean())
print(full_year_data['ride_length'].max())
```
